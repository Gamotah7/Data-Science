```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
ibovespa = yf.Ticker("^BVSP").history(period="5y")
```

```python
ibovespa
```

```python
# GrÃ¡fico original
# o mÃ©todo rolling serve para definir intervalo de tempo
ibovespa["Close"].plot(figsize=(40, 20))
ibovespa["Close"].rolling(30).mean().plot(label="MÃ©dia 30 dias")
ibovespa["Close"].rolling(90).mean().plot(label="MÃ©dia 90 dias")

# Ajustar o tamanho da fonte dos rÃ³tulos dos eixos
plt.xlabel("Data", fontsize=30)
plt.ylabel("IBOV", fontsize=30)


# Ajustar o tamanho da fonte dos valores dos eixos
plt.tick_params(axis='x', labelsize=20)  # Eixo X
plt.tick_params(axis='y', labelsize=20)  # Eixo Y

# Ajustar o tamanho da legenda
plt.legend(fontsize=30)

# Mostrar o grÃ¡fico
plt.show()
```

```python
AÃ§Ãµes = ["ITUB4.SA", "VALE3.SA", "ABEV3.SA"]
```

```python
# Loop para criar grÃ¡ficos individuais para cada aÃ§Ã£o
for acao in AÃ§Ãµes:
    # Baixar os dados da aÃ§Ã£o usando yfinance
    dados = yf.Ticker(acao).history(period="5y")
    
    # Criar o grÃ¡fico
    plt.figure(figsize=(40, 20))
    plt.plot(dados["Close"], label=f"{acao} HistÃ³rico")
    plt.plot(dados["Close"].rolling(30).mean(), label="MÃ©dia 30 dias")
    plt.plot(dados["Close"].rolling(90).mean(), label="MÃ©dia 90 dias")
    
    # Ajustar rÃ³tulos dos eixos
    plt.xlabel("Data", fontsize=30)
    plt.ylabel("PreÃ§o", fontsize=30)
    
    # Ajustar o tamanho dos valores dos eixos
    plt.tick_params(axis='x', labelsize=20)  # Eixo X
    plt.tick_params(axis='y', labelsize=20)  # Eixo Y
    
    # Ajustar a legenda
    plt.legend(fontsize=30)
    
    # Adicionar tÃ­tulo ao grÃ¡fico
    plt.title(f"GrÃ¡fico de PreÃ§o e MÃ©dias MÃ³veis - {acao}", fontsize=35)
    
    # Mostrar o grÃ¡fico
    plt.show()
```
